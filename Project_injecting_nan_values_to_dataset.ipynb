```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_org = pd.read_csv('/content/marketing_campaign_dataset.csv')
print(df_org.shape)
df_org.head()
```

```python
df_org.isnull().sum()
```

```python
df = df_org.copy()
```

```python
print("Before injecting NaN values:")
display(df.isna().sum())
print()
```

```python
num_injections = 5000
```

```python
for _ in range(num_injections):
    row_idx = np.random.randint(0, df.shape[0])
    col_idx = np.random.randint(0, df.shape[1])
    df.iat[row_idx, col_idx] = np.nan
```

```python
print("After injecting NaN values:")
display(df.isna().sum())
```

```python
df.to_csv("Marketing_campaign_dataset_with_NaN.csv", index=False)
```
