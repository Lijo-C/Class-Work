Correlation Analysis


Correlation analysis is a statistical method used to evaluate the strength and direction of a linear relationship between two quantitative variables. A correlation coefficient, typically Pearson's r, is calculated to quantify this relationship. The value of the correlation coefficient ranges from -1 to +1:

+1 indicates a perfect positive linear relationship (as one variable increases, the other increases proportionally).
-1 indicates a perfect negative linear relationship (as one variable increases, the other decreases proportionally).
0 indicates no linear relationship between the two variables.
Correlation analysis helps in understanding how variables move together, but it's important to remember that correlation does not imply causation.

Based on the correlation matrix, here are some conclusions about the relationships between the features and customer churn:

Age: Age has the highest positive correlation with churn (0.285), suggesting that older customers are more likely to churn. This aligns with the earlier histogram visualization.
Balance (Euros): There is a positive correlation between balance and churn (0.119), indicating that customers with higher balances are slightly more likely to churn.
Is Active: Being an active customer has a negative correlation with churn (-0.156), which is expected as active customers are less likely to leave.
Number of Products: There is a small negative correlation between the number of products and churn (-0.048), suggesting customers with more products might be slightly less likely to churn, but the relationship is weak.
Credit Score, Years as Customer, Has Credit Card, and Salary (Euros): These features show very weak correlations with churn (close to zero), suggesting they have little to no linear relationship with whether a customer churns or not.
Remember that correlation does not imply causation, but these insights can be helpful in understanding which factors might be associated with customer churn.
