```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
df_2025 = df[df['year'] == 2025]
```

```python
plt.figure(figsize=(6,4))
plt.bar(df_2025['retailer'], df_2025['online_revenue_million'], color=['#377eb8','#ff7f00','#4daf4a'])
plt.ylabel('Online Revenue (Million â‚¬)')
plt.title('Online Revenue by Retailer in 2025')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(df_2025['online_revenue_million'], labels=df_2025['retailer'], autopct='%1.1f%%', colors=['#377eb8','#ff7f00','#4daf4a'])
plt.title('Revenue Share in 2025')
plt.show()
```

```python
plt.figure(figsize=(8,4))
for retailer in df['retailer'].unique():
    df_r = df[df['retailer'] == retailer]
    plt.plot(df_r['year'], df_r['conversion_rate_percent'], marker='o', label=retailer)
```

```python
plt.figure(figsize=(8,4))
for retailer in df['retailer'].unique():
    df_r = df[df['retailer'] == retailer]
    plt.plot(df_r['year'], df_r['conversion_rate_percent'], marker='o', label=retailer)

plt.xlabel('Year')
plt.ylabel('Conversion Rate (%)')
plt.title('Conversion Rate Trend (2021-2025)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10,5))
for retailer in df['retailer'].unique():
    df_rt = df[df['retailer'] == retailer]
    plt.plot(df_rt['year'], df_rt['conversion_rate_percent'], marker='o', label=retailer)

plt.xlabel('Year')
plt.ylabel('Conversion Rate (%)')
plt.title('Conversion Rate Trend (2021 - 2025)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.bar(df_2025['retailer'], df_2025['website_sessions_million'], color=['#377eb8', '#ff7f00', '#4daf4a'])
plt.ylabel('Website Sessions (Million)')
plt.title('Website Sessions by Retailer (July 2025)')
plt.tight_layout()
plt.show()
```
